```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set()
```

```python
def linear(X, b0, b1):
    return [b0+b1*x for x in X]
```

```python
# b0 - Intercept
def intercept(X, Y, b1):
    x_ = np.mean(X)
    y_ = np.mean(Y)
    return y_-b1*x_
# b1 - Slope
def slope(X, Y): 
    x_ = np.mean(X)
    y_ = np.mean(Y)
    rise = sum([(x-x_) * (y-y_) for x,y in zip(X,Y)])
    run = sum([(x-x_)**2 for x,y in zip(X,Y)])
    return rise / run
```

```python
x = np.array([4,4.5,5,5.5,6,6.5,7])
```

```python
y = np.array([33,42,45,51,53,61,62])
```

```python
plt.figure(figsize=(8,5))
plt.title("x vs y")
plt.scatter(x, y, color = "#247ba0")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Predictions
#target = y
#predictor = x
b1 = slope(x, y)
b0 = intercept(x, y, b1)
predicted = linear(x, b0, b1)
plt.figure(figsize = (8, 5))
plt.plot(x, predicted, color = '#f25f5c')
plt.scatter(x, predicted, color = '#f25f5c')
plt.title('Predicted values by Linear Regression', fontsize = 15)
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y, color = "#247ba0")
plt.show()
```

```python
print("Coefficients:")
print("b0 : ", b0)
print("b1 : ", b1)
```

```python
#SSE
def r_squared(Y, Y_HAT):
    ssr, sse, r_sqr = [0]*3
    y_ = np.mean(Y)
    sse = sum([(y - y_hat)**2 for y,y_hat in zip(Y, Y_HAT)])
    sst = sum([(y - y_)**2 for y in Y])
    r_sqr = 1 - (sse / sst)
    return r_sqr
r_squared(y, predicted)
```
